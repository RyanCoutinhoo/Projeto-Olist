Neste notebook vamos realizar a extração dos dados em arquivos CSV e salva-los em formato Delta.

```python
#Onde nossos dados brutos estão armazenados
volume_source_path = "/Volumes/projeto_olist/olist/data"

#Tabela destino
target_schema = "projeto_olist.olist"

#Nome das tabelas que vamos salvar
tabelas = [
    "olist_customers_dataset",
    "olist_geolocation_dataset",
    "olist_order_items_dataset",
    "olist_order_payments_dataset",
    "olist_order_reviews_dataset",
    "olist_orders_dataset",
    "olist_products_dataset",
    "olist_sellers_dataset"
]

for tabela in tabelas:
    try:
        csv_file_path = f"{volume_source_path}/{tabela}.csv"

        #Lendo arquivo csv utilizando Spark

        df = spark.read \
            .format("csv") \
            .option("header", "true") \
            .option("inferSchema", "true") \
            .load(csv_file_path)

        target_table_name = f"{target_schema}.{tabela}_raw"

        #Gravando os dados como tabela Delta estruturada

        df.write \
            .mode("overwrite") \
            .format("delta") \
            .saveAsTable(target_table_name)
        
        print(f"Arquivo {tabela}.csv foi processado com sucesso e salvo como tabela Delta")

    except Exception as e:
        print(f"Ocorreu um erro ao processar o arquivo {tabela}.csv: {str(e)}")
```
